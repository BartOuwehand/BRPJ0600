```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import wcs
from astropy.io import ascii
from astropy.table import Table, Column, MaskedColumn
from pathlib import Path
from math import trunc
from scipy import optimize
import copy
%matplotlib inline


SMALL_SIZE = 10*2                                        
MEDIUM_SIZE = 12*2
BIGGER_SIZE = 14*2

plt.rc('font', size=SMALL_SIZE, family='serif')          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)                     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)                    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE, direction='in')    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE, direction='in')    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)                    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)                  # fontsize of the figure title
plt.rc('figure', figsize='8, 6')                         # size of the figure, used to be '4, 3' in inches
```

```python
mag = {}
magerr = {}
epoch = {}

LCOGT_raw = {}
for f in ['gp','rp','ip']:
    LCOGT_raw[f] = ascii.read("../J0600_mag/LCOGT_"+f+".txt")

ASTEP_raw = ascii.read("../J0600_mag/ASTEP_R.txt")

for f in ['gp','rp','ip']:
    mag[f] = np.array(LCOGT_raw[f]['MAG'])
    magerr[f] = np.array(LCOGT_raw[f]['MAGERR'])
    epoch[f] = np.array(LCOGT_raw[f]['EPOCH'])

mag['R'] = np.array(ASTEP_raw['MAG'])
magerr['R'] = np.array(ASTEP_raw['MAGERR'])
epoch['R'] = np.array(ASTEP_raw['EPOCH'])

filt = ['gp','rp','ip', 'R']
plot_c = {'gp':'green','rp':'r','ip':'indianred', 'R':'r'}
```

```python
fig, axs = plt.subplots(4, sharex=True, gridspec_kw={'hspace': 0})
#fig.title('Magnitude of J0600 in gp and rp filter ')
fig.set_figheight(12)
fig.set_figwidth(16)
axs[0].set_title('Relative magnitude of ASASSN-V J060000.76-310027.8 in gp, ip and rp filter')
for i in range(len(filt)):
    #if filt[i] != 'ip':
    #    axs[i].plot([0,1e6],[J0600_gaiamag[filt[i]],J0600_gaiamag[filt[i]]],c='black',linestyle='dashed')
    #axs[i].plot([0,1e6],[ZP[filt[i]],ZP[filt[i]]],c='black',linestyle='dashed')
    axs[i].errorbar(epoch[filt[i]], mag[filt[i]], yerr=magerr[filt[i]], fmt='.',c=plot_c[filt[i]])
    axs[i].set_ylabel(filt[i])
    axs[i].invert_yaxis()
    #measurement period astep telescope
axs[i].set_xlabel("Epoch[MJD]")
axs[i].set_xlim([epoch['gp'][0]-2,epoch['gp'][-1]+2])
#axs[i].set_xlim([59203,59215])


# Hide x labels and tick labels for all but bottom plot.
for ax in axs:
    ax.label_outer()
plt.show()
```
